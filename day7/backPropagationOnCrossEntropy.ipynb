```python
import numpy as np
class Loss:

    def calculate(self,output,y):
       #output-y_pred,y-y_true
       sample_losses=self.forward(output,y)
       #calls forward function from cross entropy class
       mean_loss=np.mean(sample_losses)
       #mean of loss
       return mean_loss

class Categorical_CrossEntropyLoss(Loss):
    def forward(self, y_pred, y_true):
        samples = len(y_pred)
        
        # Clip predictions to prevent log(0) or division by zero
        y_pred_clipped = np.clip(y_pred, 1e-7, 1 - 1e-7)
        
        # Case 1: y_true is class indices (1D array, e.g., [2, 0, 1])
        #if y_true(true output) is one dimensional(case 1:class targets are numbers)
        if len(y_true.shape) == 1:
            correct_confidences = y_pred_clipped[range(samples), y_true]
        
        # Case 2: y_true is one-hot encoded (2D array)
        #if y_true(true output) is 2-dimensional(case 2:class targets are one-hot encoding)
        elif len(y_true.shape) == 2:
            correct_confidences = np.sum(y_pred_clipped * y_true, axis=1)
        
        
        
        # Negative log likelihood
        negative_log_likelihoods = -np.log(correct_confidences)
        
        
        
        return negative_log_likelihoods
```
